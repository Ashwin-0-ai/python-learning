```python
#1. Setting up working directory 
import sqlite3
import pandas as pd 
import os

os.chdir("D:\\Mission_Blitzkreig\\Month_2_SQL")

conn = sqlite3.connect('marketing_campaigns.db')
cursor = conn.cursor()

print ("Working directory has been established")
print ("Connection has been estanlished to the database")
print ("Ready to execute sql queries")
```

```python
# 2 . Import data table from the database 
query = """
SELECT
      COUNT(*) as total_campaigns
FROM
      campaigns
"""
df_count = pd.read_sql_query(query,conn)

print (" Totla number of campaigns in the db is :")
print (df_count)
```

```python
# 3 . Count with where 
query = """
SELECT
       COUNT (*) campaigns_in_Mumbai
FROM campaigns
WHERE city = 'Mumbai'
"""

df_count_mumbai = pd.read_sql_query(query,conn)

print ("Total number of campaigns in mumbai is :")
print (df_count_mumbai)
```

```python
# 4. Sum function - Total Revenue  & WHERE Function
query = """
SELECT
    SUM(revenue) as total_revenue
FROM campaigns
WHERE city = 'Delhi'
"""
df_sum_revenue = pd.read_sql_query(query,conn)

print ("Total revenue generated from all the campaigns is :")
print (df_sum_revenue)
```

```python
# 5. Average Values with Where conditions
query = """
SELECT
    AVG(impressions) as average_impressions
FROM campaigns
WHERE city = "Mumbai"
"""

df_avg_impressions = pd.read_sql_query(query,conn)

print("Average impressions generated from all the campaigns is : ")
print(df_avg_impressions)
```

```python
#7. MIN() and MAX() functions 
query = """
SELECT
    MIN(revenue) as min_revenue,
    MAX(revenue) as max_revenue,
    MIN(spend) as min_spend,
    MAX(spend) as max_spend
FROM campaigns
"""

df_min_max = pd.read_sql_query(query,conn)

min_revenue = df_min_max.iloc[0]['min_revenue']
max_revenue = df_min_max.iloc[0]['max_revenue']
min_spend = df_min_max.iloc[0]['min_spend']
max_spend = df_min_max.iloc[0]['max_spend']

print ("Minimum Revenue genrated :", min_revenue)
print("Maximum revenue generated :", max_revenue)
print("Minimum spend :", min_spend)
print("Maximum Spend :", max_spend)
```

```python
# Combine all the aggregator function
query = """
SELECT
    COUNT(*) as total_campaigns,
    SUM(revenue) as total_revenue,
    AVG(impressions) as average_impressions,
    MIN(revenue) as min_revenue,
    MAX(revenue) as max_revenue,
    MIN(spend) as min_spend,
    MAX(spend) as max_spend
FROM campaigns
"""

df_agg = pd.read_sql_query(query, conn)

print ("The combined result of all the aggregators is  :")
print(df_agg)

print("Total number of campaigns in DB are :", df_agg.iloc[0]['total_campaigns'])
print("Total revenue generated : ", df_agg.iloc[0]['total_revenue'])
print("Averge impressions across the campaigns :", df_agg.iloc[0]['average_impressions'])
print("Minimum revnue generated :", df_agg.iloc[0]['min_revenue'])
print("Maximum revenue generated :", df_agg.iloc[0]['max_revenue'])
print("Minimum spend :", df_agg.iloc[0]['min_spend'])
print("Maximum spend :",df_agg.iloc[0]['max_spend'])
```

```python
#9.# Combine all the aggregator function with groupby cities  

query = """
SELECT
    City as city_name,
    COUNT(*) as total_campaigns_city,
    SUM(revenue) as total_revenue_city,
    AVG(impressions) as average_impressions_city,
    MIN(revenue) as min_revenue_city,
    MAX(revenue) as max_revenue_city,
    MIN(spend) as min_spend_city,
    MAX(spend) as max_spend_city
FROM campaigns
GROUP BY city
"""

df_agg_city = pd.read_sql_query(query, conn)

print ("The combined result of all the aggregators is  :")
print(df_agg_city)

print("Total number of campaigns per city in DB are :", df_agg_city.iloc[0]['total_campaigns_city'])
print("Total revenue generated : ", df_agg_city.iloc[0]['total_revenue_city'])
print("Averge impressions across the campaigns :", df_agg_city.iloc[0]['average_impressions_city'])
print("Minimum revnue generated :", df_agg_city.iloc[0]['min_revenue_city'])
print("Maximum revenue generated :", df_agg_city.iloc[0]['max_revenue_city'])
print("Minimum spend :", df_agg_city.iloc[0]['min_spend_city'])
print("Maximum spend :",df_agg_city.iloc[0]['max_spend_city'])
```

```python
# 10 . Groupby and Where function
query = """
SELECT
    city,
    COUNT(*) as total_campaign_city,
    sum(revenue) as total_revenue_city,
    AVG(impressions) as average_impressions_city
FROM campaigns
WHERE revenue > 5000
GROUP BY city
"""

df_filtered_agg = pd.read_sql_query(query,conn)

print ("Campaigns in the city with revenue upwards of 5000:")
print (df_filtered_agg)
```

```python
# 11: HAVING clause - filter AFTER grouping
query = """
SELECT 
    city,
    COUNT(*) as campaigns_count,
    SUM(revenue) as total_revenue,
    AVG(revenue) as average_revenue
FROM campaigns
GROUP BY city
HAVING SUM(revenue) < 100000
"""

df_high_revenue_cities = pd.read_sql_query(query, conn)

print("TOP REVENUE CITIES (over 100,000):")
print(df_high_revenue_cities)

if len(df_high_revenue_cities) > 0:
    print(f"\n✅ {len(df_high_revenue_cities)} cities generated over 100,000")
    # Count qualifying cities
else:
    print("\nNo cities generated over 100,000")
```

```python
# 12 . ORDER BY Aggregators 
query = """
SELECT
      city,
      COUNT(*) as total_campaigns_city,
      SUM(revenue) as total_revenue,
      AVG(revenue) as average_revenue,
      ROUND(AVG(revenue),2) as average_revnue_rounded
FROM campaigns 
GROUP BY city 
ORDER BY total_revenue DESC
"""

df_ranked_cities = pd.read_sql_query(query,conn)

print("CITIES RANKED BY REVENUE:")
print(df_ranked_cities)

print("\n📊 TOP 3 CITIES:\n")

# Get top 3 cities
top_3 = df_ranked_cities.head(3)
# Get first 3 rows

# Loop with proper enumeration
for position, (idx, row) in enumerate(top_3.iterrows(), start=1):
    # enumerate(start=1) gives us position 1, 2, 3 instead of 0, 1, 2
    
    print(f"{position}. {row['city']}")
    # Print rank number and city name
    
    print(f"   Campaigns: {row['total_campaigns']}")
    # Print campaign count
    
    print(f"   Total Revenue: €{row['total_revenue']:,.2f}")
    # Print total revenue
    
    print(f"   Average per Campaign: €{row['avg_revenue_rounded']:,.2f}")
    # Print average revenue
    
    print()
    # Blank line for readability
```

```python
# Cell 13: Real marketing analytics - ROI report
query = """
SELECT 
    city,
    COUNT(*) as total_campaigns,
    SUM(revenue) as total_revenue,
    SUM(spend) as total_spent,
    ROUND(AVG(revenue - spend), 2) as average_profit_per_campaign,
    ROUND((SUM(revenue) - SUM(spend)) * 100.0 / SUM(spend), 2) as roi_percentage
FROM campaigns
GROUP BY city
ORDER BY roi_percentage DESC
"""
# This calculates ROI metrics for each city:
# - total_campaigns = count of campaigns
# - total_revenue = sum of revenue
# - total_spent = sum of spend (what we spent)
# - average_profit_per_campaign = how much profit on average
# - roi_percentage = (revenue - spend) / spend * 100

df_roi_report = pd.read_sql_query(query, conn)
# Execute query

print("📊 ROI PERFORMANCE REPORT BY CITY:")
print("=" * 80)
print(df_roi_report.to_string(index=False))
# Print full report

print("\n🏆 BEST ROI CITY:")
best_roi = df_roi_report.iloc[0]
# Get top performing city

print(f"City: {best_roi['city']}")
print(f"ROI: {best_roi['roi_percentage']:.2f}%")
print(f"Total Spent: €{best_roi['total_spent']:,.2f}")
print(f"Total Revenue: €{best_roi['total_revenue']:,.2f}")
print(f"Profit: €{best_roi['total_revenue'] - best_roi['total_spent']:,.2f}")
```
