```python
import sqlite3
import pandas as pd 
import os 

os.chdir(r'D:\Mission_Blitzkreig\Month_2_SQL')
conn = sqlite3.connect('marketing_campaigns.db')
cursor = conn.cursor()

print(" Db created : marketing__campaigns")
print("Connection established")
```

```python
# Create a table 
create_table_sql = """
CREATE TABLE IF NOT EXISTS campaigns (
campaign_id INTEGER PRIMARY KEY,
campaign_name TEXT NOT NULL,
city TEXT,
format TEXT,
spend REAL,
impressions INTEGER,
clicks INTEGER,
conversions INTEGER,
revenue REAL
)
"""

cursor.execute (create_table_sql)
conn.commit()
print("Table created")
```

```python
# Load your OOH campaigns CSV
df_campaigns = pd.read_csv(r'D:\Mission_Blitzkreig\Month_1_Py\ooh_campaigns.csv')

# Add campaign data to database
df_campaigns.to_sql(
    'campaigns',           # Table name
    conn,                  # Database connection
    if_exists='replace',   # Replace table if it exists
    index=False            # Don't save the DataFrame index
)

print(f"✅ Inserted {len(df_campaigns)} campaigns into database!")
```

```python
# SQL Query 
select_sql = "SELECT * FROM campaigns"
cursor.execute(select_sql)
results = cursor.fetchall()
print ("Campaigns in db")
print ("="*80)
for row in results:
    print (row)

print(f"\nTotal campaigns: {len(results)}")
```
