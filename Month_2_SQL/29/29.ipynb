```python
# CRUD OPERATIONS

print("\n" + "="*60)
print("CRUD OPERATIONS: The Four Pillars of Database Management")
print("="*60)

crud_explanation = """
CRUD = Create, Read, Update, Delete

1️⃣  CREATE (INSERT)
   Purpose: Add new data to database
   SQL: INSERT INTO table (column1, column2) VALUES (?, ?)
   Python: cursor.execute(insert_query, data)
   
   Example: Add a new campaign to campaigns table
   ✅ Use when: Creating new records (new campaign, new manager, etc.)

2️⃣  READ (SELECT)
   Purpose: Retrieve data from database
   SQL: SELECT * FROM table WHERE condition
   Python: cursor.execute(select_query)
         results = cursor.fetchall()  OR  pd.read_sql_query()
   
   Example: Get all campaigns by a specific manager
   ✅ Use when: Querying, analyzing, displaying data

3️⃣  UPDATE
   Purpose: Modify existing data
   SQL: UPDATE table SET column = ? WHERE id = ?
   Python: cursor.execute(update_query, (new_value, record_id))
   
   Example: Update campaign budget after approval
   ✅ Use when: Changing data (status updates, corrections, etc.)

4️⃣  DELETE
   Purpose: Remove data from database
   SQL: DELETE FROM table WHERE condition
   Python: cursor.execute(delete_query, (record_id,))
   
   Example: Delete a cancelled campaign
   ✅ Use when: Removing obsolete data
   ⚠️  CAREFUL: Deletes are permanent!

---

TRANSACTION MANAGEMENT:
- conn.commit() → Save all changes permanently
- conn.rollback() → Undo changes if something goes wrong
- Used together to ensure all-or-nothing operations
"""

print(crud_explanation)
print("\n" + "="*60)
```

```python
import sqlite3
import pandas as pd
import os

os.chdir(r'D:\Mission_Blitzkreig\Month_2_SQL')

with sqlite3.connect('marketing_campaigns_normalized.db') as conn:
    
    # READ 1: All campaigns
    query1 = """
    SELECT 
        campaign_id,
        campaign_name,
        budget,
        actual_spend
    FROM campaigns
    ORDER BY campaign_id
    """
    df_all = pd.read_sql_query(query1, conn)
    
    # READ 2: With JOIN
    query2 = """
    SELECT 
        c.campaign_id,
        c.campaign_name,
        ci.city_name,
        cm.manager_name,
        cf.format_name,
        c.actual_spend
    FROM campaigns c
    INNER JOIN cities ci ON c.city_id = ci.city_id
    INNER JOIN campaign_managers cm ON c.manager_id = cm.manager_id
    INNER JOIN campaign_formats cf ON c.format_id = cf.format_id
    ORDER BY c.campaign_id
    """
    df_join = pd.read_sql_query(query2, conn)
    
    # READ 3: Aggregated
    query3 = """
    SELECT 
        ci.city_name,
        COUNT(c.campaign_id) as campaign_count,
        SUM(c.actual_spend) as total_spend,
        AVG(c.conversions) as avg_conversions
    FROM cities ci
    LEFT JOIN campaigns c ON ci.city_id = c.city_id
    GROUP BY ci.city_id, ci.city_name
    ORDER BY total_spend DESC
    """
    df_agg = pd.read_sql_query(query3, conn)

print(f"✅ df_all: {len(df_all)} records")
print(f"✅ df_join: {len(df_join)} records")
print(f"✅ df_agg: {len(df_agg)} records")

print("\n📊 City Summary:")
print(df_agg.to_string(index=False))
```

```python
# method 2 : Select with WHERE clause 
from csv import writer


print("\n Read with filtering with Where condition")

query_filtered = """ 
SELECT
     campaign_id,
     campaign_name,
     budget,
     (budget - actual_spend) as remaining_budget
FROM campaigns
WHERE budget > 30000
ORDER BY actual_spend DESC
"""

df_filtered = pd.read_sql_query(query_filtered,conn)
print(f"Found {len(df_filtered)} campaigns with budget > 30000")
print(df_filtered.to_string(index=False))

# method 3  : SELECT with Joins 
query_join = """ 
SELECT 
      c.campaign_id,
      c.campaign_name,
      ci.city_name,
      cm.manager_name,
      cf.format_name,
      c.actual_spend
FROM campaigns c
INNER JOIN cities ci ON c.city_id = ci.city_id
INNER JOIN campaign_managers cm ON c.manager_id = cm.manager_id
INNER JOIN campaign_formats cf on c.format_id = cf.format_id
ORDER BY c.campaign_id
"""

df_join = pd.read_sql_query(query_join,conn)
print(f"Retrieved {len(df_join)} campaigns with JOINS")
print(df_join.to_string(index=False))

# method 4  : SELECT with Aggregations 

print ("\n Select with Aggregations")
query_agg = """
SELECT 
     ci.city_name ,
     COUNT(c.campaign_id) as total_campaigns,
     SUM (c.actual_spend) as total_spend,
     AVG(c.conversions) as avg_conversions,
     MAX(c.budget) as max_budget
FROM cities ci
LEFT JOIN campaigns c ON ci.city_id = c.city_id
GROUP BY ci.city_id,ci.city_name
HAVING COUNT(c.campaign_id) > 0 
ORDER BY total_spend DESC
"""

# method 5 :Export to CSV
print("\n Exporting to CSV")

csv_file ='campaign_export.csv'
df_join.to_csv(csv_file,index=False)
print(f"✅ Exported {len(df_join)} records to {csv_file}")       

# method 6 : Export to Excel 
print ("\n Exporting to Excel")
excel_file = 'campaign_export.xlsx'
with pd.ExcelWriter(excel_file, engine='openpyxl') as writer:
        df_join.to_excel(writer, sheet_name='Campaigns', index=False)
        df_agg.to_excel(writer, sheet_name='Summary', index=False)

print(f"Exported data to {excel_file}")
print(f"   - Sheet 1: Campaign details")
print(f"   - Sheet 2: City summary")
```

```python
# 5. Update Operations 
print ("UPDATE OPERATIONS")
with sqlite3.connect('marketing_campaigns_normalized.db') as conn :
    cursor = conn.cursor()

# Method 1 : Update single record 
print("\n Before Update:")

    # Show the current data   
query_before = "SELECT campaign_id,campaign_name,actual_spend FROM campaigns WHERE campaign_id = 1"
df_before = pd.read_sql_query(query_before, conn)
print(df_before.to_string(index=False))

    # Update the Record 
update_query = """
UPDATE campaigns
SET actual_spend = ?
WHERE campaign_id = ?
"""
cursor.execute(update_query,(48000,1))
conn.commit()

print("\n After Update :")
df_after = pd.read_sql_query(query_before, conn)
print(df_after.to_string(index=False))
print(f"Update campaign 1 spend to 48000")

# Method 2 : Update Multiple Records 
print("\n Update Multiple Records")
update_multiple = """ 
UPDATE campaigns
SET conversions = conversions + 10
WHERE manager_id = 1 
"""

cursor.execute(update_multiple)
rows_updated = cursor.rowcount 
conn.commit()

print(f"Updated {rows_updated} campaigns")
print(f"Incremented conversions by 10 for manager 1's campaigns")

# Method 3 : Updated with calculated Value 
print("\n3️⃣  UPDATE WITH CALCULATED VALUES")
print("-" * 60)
    
update_calc = """
    UPDATE campaigns
    SET clicks = ROUND((actual_spend / 100) * 5)  -- Calculate clicks based on spend
    WHERE campaign_id = 2
    """
    
cursor.execute(update_calc)
conn.commit()

print(f"✅ Updated campaign 2 clicks based on spend")
    
    # METHOD 4: Update with error handling
print("\n4️⃣  UPDATE WITH ERROR HANDLING")
print("-" * 60)
    
try:
        update_safe = """
        UPDATE campaigns
        SET manager_id = ?
        WHERE campaign_id = ?
        """
        
        cursor.execute(update_safe, (99, 1))  
        conn.commit()
        print("✅ Campaign updated")
except sqlite3.IntegrityError as e:
        print(f"⚠️  Error: {e}")
        print(f"⚠️  Manager ID 99 doesn't exist (FOREIGN KEY violation)")
        conn.rollback()
    
    # Show final state
print("\n📊 Final campaign data:")
query_final = """
    SELECT campaign_id, campaign_name, actual_spend, clicks, conversions
    FROM campaigns
    ORDER BY campaign_id
    """
df_final = pd.read_sql_query(query_final, conn)
print(df_final.to_string(index=False))

print("\n✅ UPDATE operations completed!")



```

```python
print("\n" + "="*60)
print("PART 6: DELETE OPERATION")
print("="*60)

print("\n⚠️  WARNING: DELETE IS PERMANENT!")
print("Always backup before deleting!")

with sqlite3.connect('marketing_campaigns_normalized.db') as conn:
    cursor = conn.cursor()
    
    # First, insert a test record to delete
    print("\n1️⃣  INSERT TEST RECORD")
    print("-" * 60)
    
    insert_test = """
    INSERT INTO campaign_formats (format_name, description, average_cpm)
    VALUES (?, ?, ?)
    """
    
    cursor.execute(insert_test, ('Test Format', 'Testing only', 0.00))
    test_id = cursor.lastrowid
    conn.commit()
    
    print(f"✅ Inserted test format with ID: {test_id}")
    
    # METHOD 1: Delete single record
    print("\n2️⃣  DELETE SINGLE RECORD")
    print("-" * 60)
    
    print("BEFORE delete:")
    query_before = """
    SELECT format_id, format_name FROM campaign_formats
    ORDER BY format_id DESC LIMIT 1
    """
    df_before = pd.read_sql_query(query_before, conn)
    print(df_before.to_string(index=False))
    
    delete_query = """
    DELETE FROM campaign_formats
    WHERE format_id = ?
    """
    
    cursor.execute(delete_query, (test_id,))
    deleted_rows = cursor.rowcount  
    conn.commit()
    
    print(f"\n✅ Deleted {deleted_rows} test format record")
    
    # METHOD 2: Delete with condition
    print("\n3️⃣  DELETE WITH CONDITION")
    print("-" * 60)
    
    # Count how many would be deleted
    count_query = """
    SELECT COUNT(*) as count
    FROM campaigns
    WHERE actual_spend < 10000  -- Campaigns that spent less than 10k
    """
    
    df_count = pd.read_sql_query(count_query, conn)
    count = df_count.iloc[0, 0]
    
    print(f"⚠️  Would delete {count} campaigns with spend < 10k")
    print("❌ NOT deleting (just for demonstration)")
    
    # METHOD 3: Safe delete with backup
    print("\n4️⃣  SAFE DELETE (WITH BACKUP)")
    print("-" * 60)
    
    # Before deleting, backup the data
    backup_query = """
    SELECT * FROM campaigns
    WHERE campaign_id = 4  -- Specific campaign to delete
    """
    
    df_backup = pd.read_sql_query(backup_query, conn)
    backup_file = 'deleted_campaigns_backup.csv'
    df_backup.to_csv(backup_file, index=False)
    print(f"✅ Backed up campaign data to {backup_file}")
    
    # Now safe to delete (data is backed up)
    print("✅ Backup complete, ready to delete if needed")
    
    # METHOD 4: Soft delete (mark as deleted instead)
    print("\n5️⃣  SOFT DELETE (Better Practice!)")
    print("-" * 60)
    
    print("""
    HARD DELETE (what we've been doing):
    DELETE FROM campaigns WHERE campaign_id = 1
    → Record is GONE permanently
    → Can't recover
    → Breaks referential integrity
    
    SOFT DELETE (better):
    UPDATE campaigns SET is_deleted = TRUE WHERE campaign_id = 1
    → Record still in database (marked as deleted)
    → Can recover if needed
    → Maintains referential integrity
    → Can restore with: UPDATE campaigns SET is_deleted = FALSE
    
    ✅ RECOMMENDATION: Use soft deletes in production!
    """)
    
    print("\n📊 Current campaigns in database:")
    query_final = "SELECT campaign_id, campaign_name FROM campaigns ORDER BY campaign_id"
    df_final = pd.read_sql_query(query_final, conn)
    print(f"✅ Total campaigns: {len(df_final)}")
    print(df_final.to_string(index=False))

print("\n✅ DELETE operations completed!")
print("⚠️  Remember: Always backup before deleting!")
```
