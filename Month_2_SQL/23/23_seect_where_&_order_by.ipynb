```python
# 1 . Get All data 
import sqlite3
import pandas as pd 
import os 

os.chdir(r'D:\Mission_Blitzkreig\Month_2_SQL')
conn = sqlite3.connect('marketing_campaigns.db')
cursor = conn.cursor()
print ("DB connected successfully")
```

```python
# 2.Basic select query 
query = "SELECT * From campaigns"
df_all = pd.read_sql_query(query,conn)
print("Total campaigns in DB :", len(df_all))

print("\n All campaigns:")
print(df_all)

print("\n Columns in campaigns table :", df_all.columns.tolist())
```

```python
#3. Where - Filter data 
query = """
SELECT
      campaign_name,
      city,
      revenue,
      spend
FROM campaigns
WHERE city = 'Mumbai'
"""
df_mumbai = pd.read_sql_query(query, conn)

print("Campaigns in Mumbai : ", len(df_mumbai))
print(df_mumbai)

print(f"\nTotal campaigns in Mumbai: {len(df_mumbai)}")
```

```python
# 4. Where with condition
query = """
SELECT
       campaign_name,
       city,
       spend
FROM campaigns
WHERE spend <5000
"""
df_spend_less_5000 =pd.read_sql_query(query, conn)
print("Campaigns with spennd less that 5000 :")
print(df_spend_less_5000)

print(f"\nCount: {len(df_spend_less_5000)}")
```

```python
# 5. Filter and Logic 
query = """
SELECT
       campaign_name,
       city,
       spend
FROM campaigns
WHERE spend < 5000 AND city = 'Delhi'
"""
df_and_func = pd.read_sql_query(query, conn)

print("Delhi with campaigns spend less than 5000 :")
print(df_and_func)

print(f"\n Count : {len(df_and_func)}")
```

```python
#6. Filter with OR conditions 
query = """
SELECT
      campaign_name,
      city,
      spend
FROM campaigns
WHERE city = 'Delhi' OR city = 'Mumbai'
"""
df_or_func =pd.read_sql_query(query,conn)

print("Campaigns in delhi or Mumbai:")
print(df_or_func)

print(f"\n Count  : {len(df_or_func)}")
```

```python
# 7: ORDER BY - ASC 
query = """
SELECT 
    campaign_name,
    city,
    revenue,
    spend
FROM campaigns
ORDER BY revenue ASC
"""

df_sorted_asc = pd.read_sql_query(query, conn)

print("📈 Campaigns sorted by revenue (LOWEST first):")
print(df_sorted_asc)


print(f"\nLowest revenue campaign: {df_sorted_asc.iloc[0]['revenue']:,.2f}")
```

```python
# 7: ORDER BY - DSC
query = """
SELECT 
    campaign_name,
    city,
    revenue,
    spend
FROM campaigns
ORDER BY revenue DESC
"""

df_sorted_dsc = pd.read_sql_query(query, conn)

print("📈 Campaigns sorted by revenue (HIGHEST first):")
print(df_sorted_dsc)


print(f"\n Highest revenue campaign: {df_sorted_dsc.iloc[0]['revenue']:,.2f}")
```

```python
# 9 : WHERE + ORDER BY - filter AND sort
query = """
SELECT 
    campaign_name,
    city,
    revenue,
    spend
FROM campaigns
WHERE city = 'Delhi' AND revenue < 20000
ORDER BY spend DESC
"""

df_filtered_sorted = pd.read_sql_query(query, conn)


print("🎯 Best performing Delhi campaigns (revenue > 20,000):")
print(df_filtered_sorted)

if len(df_filtered_sorted) > 0:
    best_campaign = df_filtered_sorted.iloc[0]
    print(f"\n BEST CAMPAIGN:")
    print(f"  Name: {best_campaign['campaign_name']}")
    print(f"  ROI: {best_campaign['roi']:.2f}%")
    print(f"  Revenue: €{best_campaign['revenue']:,.2f}")
else:
    print("\nNo campaigns match these criteria")
```
