```python
import pandas as pd

# Create sample OOH campaign data
campaign_data = """city,format,impressions,clicks,cost_eur,revenue_eur,duration_days
Berlin,Billboard,150000,450,2500,5400,30
Berlin,Transit,120000,360,1800,4320,30
Munich,Billboard,100000,250,2000,3750,30
Munich,Digital,80000,320,1200,3840,30
Hamburg,Billboard,90000,180,1800,2700,30
Hamburg,Transit,110000,330,1600,3960,30
Frankfurt,Billboard,120000,360,2400,4320,30
Frankfurt,Digital,95000,380,1400,4560,30"""

# Save to current directory
with open('ooh_campaigns_new.csv', 'w') as f:
    f.write(campaign_data)

# Load it
df = pd.read_csv('ooh_campaigns_new.csv')
print(df)
```

```python
# Group by city and sum the revenue
revenue_by_city = df.groupby('city')['revenue_eur'].sum()
print("Total Revenue by City:")
print(revenue_by_city)
```

```python
# Group by city and calculate multiple metrics
city_summary = df.groupby('city').agg({
    'impressions': 'sum',
    'clicks': 'sum',
    'cost_eur': 'sum',
    'revenue_eur': 'sum'
})

print("Campaign Summary by City:")
print(city_summary)
```

```python
# Create a summary table with calculations
city_metrics = df.groupby('city').agg({
    'impressions': 'sum',
    'clicks': 'sum',
    'cost_eur': 'sum',
    'revenue_eur': 'sum'
})

# Calculate marketing metrics
city_metrics['CTR'] = (city_metrics['clicks'] / city_metrics['impressions'] * 100).round(2)
# CTR = Click-Through Rate (% of impressions that clicked)

city_metrics['CPM'] = (city_metrics['cost_eur'] / (city_metrics['impressions'] / 1000)).round(2)
# CPM = Cost Per Mille (cost per 1,000 impressions)

city_metrics['ROI'] = ((city_metrics['revenue_eur'] - city_metrics['cost_eur']) / city_metrics['cost_eur'] * 100).round(2)
# ROI = Return On Investment (profit % on spending)

print("City Performance Metrics:")
print(city_metrics[['impressions', 'clicks', 'cost_eur', 'revenue_eur', 'CTR', 'CPM', 'ROI']])

```

```python
# 1. Group by format (Billboard, Transit, Digital)

# Step 1: Group by format
grouped=df.groupby('format')

# Print the groups
for format_name, group in grouped:
    print(f"\n{format_name}:")
    print(group)
```

```python
# 2. Calculate total impressions and revenue for each
Total_impressions_revenue = df.groupby('format').agg({
                                                      'impressions':'sum',
                                                      'revenue_eur':'sum'
})

print("The total impressions & revenue byf format :")
print(Total_impressions_revenue)
```

```python
# 3. Calculate ROI for each format
roi_format = df.groupby('format').agg({'cost_eur': 'sum',
                                       'revenue_eur': 'sum'
    })

roi_format['ROI'] = ((roi_format['revenue_eur'] - roi_format['cost_eur']) / roi_format['cost_eur'] * 100).round(2)

print("ROI grouped by formats:")
print(roi_format)

```

```python
# 4. Which format has the best ROI?
best_roi = roi_format['ROI'].idxmax()
print("Format with best ROI:", best_roi)
print("ROI value:", roi_format['ROI'].max())
```
