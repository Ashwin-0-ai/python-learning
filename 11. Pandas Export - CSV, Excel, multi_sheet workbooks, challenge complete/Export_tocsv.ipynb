```python
# SETUP CELL — RUN THIS FIRST
import os
import pandas as pd

# Set working directory
os.chdir(r'D:\Mission_Blitzkreig')
print("✅ Working in:", os.getcwd())
```

```python
 # load data from day 10
df =pd.read_csv('ooh_campaigns.csv')

print(df)
```

```python
# Calculate city metrics 
city_metrics = df.groupby('city').agg({
                                        'impressions' : 'sum',
                                        'clicks': 'sum',
                                        'conversions' : 'sum',
                                        'revenue' : 'sum',
                                        'spend' : 'sum'
})
```

```python
# Calcuate KPI

city_metrics['ctr'] = (city_metrics['clicks'] / city_metrics['impressions'] * 100).round(2)

city_metrics['cpm'] = (city_metrics['spend'] / city_metrics['impressions'] * 1000).round(2)

city_metrics['roi'] = ((city_metrics['revenue'] - city_metrics['spend'] ) / city_metrics['spend'] * 100).round(2)

# Export to CSV 

city_metrics.to_csv('city_perfrormance_report.csv')

print ( "CSV SAVED : city_perfrormance_report.csv")

```

```python
# Export it to excel 

city_metrics.to_excel('city_metric_report.xlsx', sheet_name='City Metrics')
print("DONE")
```

```python
# Multiple sheet in one XLS 
format_metrics = df.groupby('format').agg({
    'impressions': 'sum',
    'revenue': 'sum',
    'spend': 'sum'
})
format_metrics['ROI'] = ((format_metrics['revenue'] - format_metrics['spend']) / format_metrics['spend'] * 100).round(2)
```

```python
# Create Writer OBJ :

with pd.ExcelWriter('city_metric_report.xlsx') as writer :

# City Metrics
    city_metrics.to_excel(writer,sheet_name='By City')

# Format Metrics
    city_metrics.to_excel(writer,sheet_name='By Format')

#Raw data 

df.to_excel(writer,sheet_name='Raw data', index=False)

print(" Multiple sheet has been done ")
```

```python
# Export Specific column only 
# Export only the most important columns
summary = city_metrics[['revenue', 'spend', 'roi', 'ctr']].copy()
summary.to_excel('executive_summary.xlsx', sheet_name='Summary')
print("✅ Executive summary saved!")
```

```python
# Without reset_index, 'city' becomes the index column
city_metrics.to_excel('with_index.xlsx')

# With reset_index, 'city' becomes a regular column
city_metrics.reset_index().to_excel('without_index.xlsx', index=False)
print("✅ Clean export with city as a column!")
```

```python
# 1. CHALLENGE: Group data by BOTH city AND format
group_by_city_format = df.groupby(['city','format']).agg({
    'revenue':'sum',
    'spend':'sum'
})
print("Below are grouped by city & format ")
print(group_by_city_format)
```

```python
2. # Calculate total revenue and ROI for each combination
group_by_city_format ['roi'] = ((group_by_city_format['revenue'] - group_by_city_format['spend'])/group_by_city_format['spend']*100).round(2)
print("calculated by total revenue and ROI")
print(group_by_city_format)
```

```python
# 3. Export to Excel with TWO sheets:

# First, create city_format_metrics (groupby city AND format)
city_format_metrics = df.groupby(['city', 'format']).agg({
    'revenue': 'sum',
    'spend': 'sum'
})

# Calculate ROI
city_format_metrics['roi'] = ((city_format_metrics['revenue'] - city_format_metrics['spend']) / city_format_metrics['spend'] * 100).round(2)

# Create top 5 performers
top_performers = city_format_metrics.sort_values('roi', ascending=False).head(5)

# Now export to Excel
with pd.ExcelWriter('city_format_analysis.xlsx') as writer:
    # Sheet 1: All city + format combinations
    city_format_metrics.to_excel(writer, sheet_name='City + Format Performance')
    
    # Sheet 2: Top 5 performers only
    top_performers.to_excel(writer, sheet_name='Top 5 Performers')

print("✅ Multiple sheets exported successfully!")
```
