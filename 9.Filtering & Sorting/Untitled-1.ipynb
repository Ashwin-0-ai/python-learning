```python
import pandas as pd
import os

# Check where you are
print("Current directory:", os.getcwd())

# Create sample OOH campaign data
campaign_data = """city,format,impressions,clicks,cost_eur,revenue_eur,duration_days
Berlin,Billboard,150000,450,2500,5400,30
Berlin,Transit,120000,360,1800,4320,30
Munich,Billboard,100000,250,2000,3750,30
Munich,Digital,80000,320,1200,3840,30
Hamburg,Billboard,90000,180,1800,2700,30
Hamburg,Transit,110000,330,1600,3960,30
Frankfurt,Billboard,120000,360,2400,4320,30
Frankfurt,Digital,95000,380,1400,4560,30"""

# Save to current directory (simple)
with open('ooh_campaigns.csv', 'w') as f:
    f.write(campaign_data)

# Load it
df = pd.read_csv('ooh_campaigns.csv')
print("✅ SUCCESS! CSV loaded:")
print(df)
```

```python
# Show only Berlin campaigns 
berlin_campaigns =df[df['city'] == 'Berlin']
print ("Berlin campaigns:")
print(berlin_campaigns)
```

```python
# Filter : Berlin & Billboard 
berlin_billboards = df[(df['city'] == 'Berlin') & (df['format'] == 'Billboard')]
print ("Berlin Billboard:")
print(berlin_billboards)
```

```python
# Filter Berlin & Munich 
berlin_munich = df[(df['city'] == 'Berlin') | (df['city'] == 'Munich')]
print ("Berlin & Munich:")
print(berlin_munich)
```

```python
# Filter revenue > 400 Eur 
high_revenue =df[df['revenue_eur'] > 4000]
print ("Revenue > 4000 Eur:")
print(high_revenue)
```

```python
# Which are high-performing camoaigns (rev >4000 Eur) 

major_cities = ['Berlin', 'Munich', 'Hamburg', 'Frankfurt ']
high_performance = df[(df['revenue_eur'] > 4000) & (df['city'].isin(major_cities))].sort_values ('revenue_eur', ascending=False)
print ("Top performance in major cities:")
print(high_performance)
```

```python
# CHALLENGE:
# 1. Filter campaigns with MORE than 100,000 impressions
# 2. Sort them by impressions (highest first)
# 3. Print only the city, format, and impressions columns

# Challenge 1 : Filter campaigns with MORE than 100,000 impressions
high_impressions = df[df['impressions'] > 100000]
print ("Campaign with more than 100000 impressons:")
print(high_impressions)
```

```python
# Challenge 2 : Sort them by impressons (highest first)
high_impressions_sorted = high_impressions.sort_values('impressions', ascending=False)
print ("Campaign with more than 100000 imprssons sorted:")
print (high_impressions_sorted)
```

```python
# Challenge 3 : print only the city , format and impressions columns 
high_impressions_sorted_selected = high_impressions_sorted[['city', 'format', 'impressions']]
print ("Campaign with more than 100000 imprssons sorted and selected columns:") 
print (high_impressions_sorted_selected)
```

```python
result = df[df['impressions'] > 100000]\
           .sort_values('impressions', ascending=False)\
           [['city', 'format', 'impressions']]
print(result)
```
